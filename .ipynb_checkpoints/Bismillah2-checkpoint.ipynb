```python
import pandas as pd

features = pd.read_csv('dummy/dummy_Hasil_Tf_Idf.csv')
features
```

```python
print('jumlah data dan kolom:', features.shape)
```

```python
features.describe()
```

```python
labels_read = pd.read_csv('dummy/dummy_label.csv')
labels_read
```

```python
import numpy as np

labels = np.array(labels_read['Politik'])
print(labels)
```

```python
feature_list = list(features.columns)
features = np.array(features)
print(features)
```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
print('Training Features:', train_features)
print('Training Labels:', train_labels)
print('Testing Features:', test_features)
print('Testing Labels:', test_labels)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 50, criterion="entropy")
classifier.fit(train_features, train_labels)
classifier_predictions = classifier.predict(test_features)

print(classifier_predictions , ' --- ', test_labels)
```

```python
# # Calculate the absolute errors
# errors = abs(classifier_predictions - test_labels)
# # Print out the mean absolute error (mae)
# print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot
# Pull out one tree from the forest
tree = classifier.estimators_[10]
# Export the image to a dot file
export_graphviz(tree, out_file = 'graph/tree.dot', feature_names = feature_list, rounded = True, precision = 1)
# Use dot file to create a graph
(graph, ) = pydot.graph_from_dot_file('graph/tree.dot')
# Write graph to a png file
graph.write_png('graph/tree.png')
```

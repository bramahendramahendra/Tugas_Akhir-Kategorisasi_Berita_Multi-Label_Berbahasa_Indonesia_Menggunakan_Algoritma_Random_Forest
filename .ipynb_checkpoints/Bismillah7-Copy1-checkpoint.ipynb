```python
class Regex():
    def Indonesia_Tagger(self, data, ct):
        clean_punct = RegexpTokenizer(r'\w+') 
        arr = [clean_punct.tokenize(data.lower())]
        hasil = ct.tag_sents(arr)
        return hasil
    
    # memasukkan fitur ke dictionary
    def insert_dict(self, kata, fitur):
        if kata not in fitur:
            fitur[kata] = 1
        else:
            fitur[kata] += 1
        return fitur
    
    def TF_Regex_Pos(self, data, politik, ekonomi, teknologi):
        hasil = {}
        for i, sentence in enumerate(data):
            for j, word_tag in enumerate(sentence):
                # politik
                if(word_tag[0] == 'pemilihan'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'umum'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                elif(word_tag[0] == 'calon'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'presiden'):
                        self.insert_dict(word_tag[0]+word_after, hasil) 
                    elif(word_after == 'wakil'):
                        word_after2 = sentence[j+2][0]
                        if (word_after2 == 'presiden'):
                            self.insert_dict(word_tag[0]+word_after+word_after2, hasil)
                elif(word_tag[0] == 'dualisme'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'kepengurusan'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                # politik
                elif (word_tag[0] in politik):
                    tag_of_word = word_tag[0]
                    self.insert_dict(word_tag[0], hasil)
                #ekonomi
                elif(word_tag[0] == 'pasar'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'saham'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                elif(word_tag[0] == 'pendapatan'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'negara'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                elif(word_tag[0] == 'suku'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'bunga'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                elif(word_tag[0] == 'tarif'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'pajak'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                #ekonomi
                elif (word_tag[0] in ekonomi):
                    tag_of_word = word_tag[0]
                    self.insert_dict(word_tag[0], hasil)
                #teknologi
                elif(word_tag[0] == 'perkembangan'):
                    word_after = sentence[j+1][0]
                    if (word_after == 'inovasi'):
                        self.insert_dict(word_tag[0]+word_after, hasil)
                #teknologi
                elif (word_tag[0] in teknologi):
                    tag_of_word = word_tag[0]
                    self.insert_dict(word_tag[0], hasil)
        return hasil
    
    def IDF_Regex_Pos(self, data, fitur_all_doc):
        for i, fitur in enumerate(data):
            if fitur not in fitur_all_doc:
                fitur_all_doc[fitur] = 1
            else:
                fitur_all_doc[fitur] += 1
        return fitur_all_doc
    
    def Proses_Regex(self, data, IT, ct, TFR, politik, ekonomi, teknologi, IDFR): #proses tagger
        if(IDFR == True):
            Hasil_Proses_Berita = []
            fitur_all_doc = {} 
            for berita in data.Berita:
                if(IT == True):
                    Hasil_Data = self.Indonesia_Tagger(berita, ct)
                if(TFR == True):
                    Hasil_Data = self.TF_Regex_Pos(Hasil_Data, politik, ekonomi, teknologi)
                self.IDF_Regex_Pos(Hasil_Data, fitur_all_doc)
            Hasil_Proses_Berita.append(fitur_all_doc)
            return Hasil_Proses_Berita, fitur_all_doc
        else :
            Hasil_Proses_Berita = []
            for berita in data.Berita:
                if(IT == True):
                    Hasil_Data = self.Indonesia_Tagger(berita, ct)
                if(TFR == True):
                    Hasil_Data = self.TF_Regex_Pos(Hasil_Data, politik, ekonomi, teknologi)
                Hasil_Proses_Berita.append(Hasil_Data)
            return Hasil_Proses_Berita, 0

    def Outputan_Data(self):
        for i in range(0, len(self.data_hasil['berita'])):
            print(str(i+1))
            print(self.data_hasil['berita'][i])     
            print('================================================')
        return None

    def __init__(self, data, IT, TFR, IDFR): #iniisialisasi data tagger
        ct = CRFTagger()
        ct.set_model_file('./library/tagger/indonesian_tagger')
        politik = ['politik', 'capres', 'cawapres', 'kampanye', 'kementrian', 'menteri', 'partai', 'pemerintah', 'pemerintahan',  
                   'pemilu', 'pemkot', 'pemprov', 'politikus', 'presiden', 'pemilihan umum', 'wakil presiden']
        ekonomi = ['ekonomi', 'anggaran', 'biaya', 'bisnis', 'harga', 'investasi', 'keuangan',  'pajak', 'perbankan', 
                   'perdagangan', 'saham']
        teknologi = ['teknologi', 'android', 'aplikasi', 'berinovasi','fitur', 'game', 'inovatif', 'jaringan', 'penelitian', 'perangkat', 
                     'platform', 'programer', 'smartphone', 'upgrade', 'versi']
        
        #Proses tagger
        self.fitur_one_doc, self.fitur_all_doc = self.Proses_Regex(data, IT, ct, TFR, politik, ekonomi, teknologi, IDFR)
        
        self.data_hasil = {}
        self.data_hasil['berita'] = self.fitur_one_doc
        
    def Get_Fitur_One_Doc(self):
        return self.fitur_one_doc
    
    def Get_Fitur_All_Doc(self):
        return self.fitur_all_doc
```

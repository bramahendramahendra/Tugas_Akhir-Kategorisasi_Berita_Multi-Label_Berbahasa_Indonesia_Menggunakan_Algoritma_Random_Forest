# Preprocessing

```python
class Preprocessing():
    
    def Punctuation_Removal(self, data): #Punctional removal adalah proses penghilangan tanda baca pada teks
        Hasil_Punctuation_Removal = re.sub("[^a-zA-Z]", " ", data)
        return Hasil_Punctuation_Removal
    
    def Case_Folding(self, data): #case folding adalah proses mengubah huruf besar menjadi kecil
        Hasil_Case_Folding = data.lower()
        return Hasil_Case_Folding
    
    def StopWords_Removal(self, data, stopword):
        StopWords_Removal = stopword.remove(data)
        return StopWords_Removal
        
    def Stemming_Data(self, data, stemmer):
        Hasil_Stemming_Data = stemmer.stem(data)
        return Hasil_Stemming_Data
    
    def Proses_Preprocessing(self, data, PR, CF, stopword, SWR, stemmer, SM): #proses prerocessing
        Hasil_Proses_Berita = []
        for berita in data.Berita:
            if(PR == True):
                Hasil_Data = self.Punctuation_Removal(berita)
            if(CF == True):
                Hasil_Data = self.Case_Folding(Hasil_Data)
            if(SWR == True):
                Hasil_Data = self.StopWords_Removal(Hasil_Data, stopword)
            if(SM == True):
                Hasil_Data = self.Stemming_Data(Hasil_Data, stemmer)
            Hasil_Proses_Berita.append(Hasil_Data)
        return Hasil_Proses_Berita
    
    def Save_Excel(self, berita):
        filename = 'dummy_clean'
        filepath = os.getcwd()+'/dummy/' +filename +'.csv'
        raw_data = {'Berita': berita}
        df = pd.DataFrame(raw_data, columns = ['Berita'])
        df.to_csv(filepath, index=False)
        return filename
    
    def Outputan_Data(self):
        for i in range(0, len(self.data_hasil['berita'])):
            print(str(i+1) +'.- ' +self.data_hasil['berita'][i])
            print(' ')
            print('---------------------------------------------------')
            print(' ')
        
        return None
    
    def __init__(self, data, PR, CF, SWR, SM, save): #iniisialisasi data preprocessing
        #inisialisasi StopWords
        factory_SWR = StopWordRemoverFactory()
        stopword = factory_SWR.create_stop_word_remover()
        #inisialisasi Stemming
        factory_S = StemmerFactory()
        stemmer = factory_S.create_stemmer()
        #Poses Preprocessing
        hasil_preprocessing = self.Proses_Preprocessing(data, PR, CF, stopword, SWR, stemmer, SM)
        #menyimpda data ke excel
        if(save == True):
            self.filepath = self.Save_Excel(hasil_preprocessing)
        self.data_hasil = {}
        self.data_hasil['berita'] = hasil_preprocessing
        
```

```python
class Regex():
    def Indonesia_Tagger(self, data, ct):
        clean_punct = RegexpTokenizer(r'\w+') 
        arr = [clean_punct.tokenize(data.lower())]
        hasil = ct.tag_sents(arr)
        return hasil
    
    # memasukkan fitur ke dictionary
    def insert_dict(self, kata, fitur):
        if kata not in fitur:
            fitur[kata] = 1
        else:
            fitur[kata] += 1
        return fitur
    
    def TF_Regex_Pos(self, data, politik, ekonomi, teknologi):
        hasil = {}
        for i, sentence in enumerate(data):
            for j, word_tag in enumerate(sentence):
                # politik
                if (word_tag[0] in politik):
                    tag_of_word = word_tag[0]
                    self.insert_dict(word_tag[0], hasil)
                elif (word_tag[0] in ekonomi):
                    tag_of_word = word_tag[0]
                    self.insert_dict(word_tag[0], hasil)
                elif (word_tag[0] in teknologi):
                    tag_of_word = word_tag[0]
                    self.insert_dict(word_tag[0], hasil)
        return hasil
    
    def IDF_Regex_Pos(self, data, fitur_all_doc):
        for i, fitur in enumerate(data):
            if fitur not in fitur_all_doc:
                fitur_all_doc[fitur] = 1
            else:
                fitur_all_doc[fitur] += 1
        return fitur_all_doc
    
    def Proses_Regex(self, data, IT, ct, TFR, politik, ekonomi, teknologi, IDFR): #proses tagger
        if(IDFR == True):
            Hasil_Proses_Berita = []
            fitur_all_doc = {} 
            for berita in data.Berita:
                if(IT == True):
                    Hasil_Data = self.Indonesia_Tagger(berita, ct)
                if(TFR == True):
                    Hasil_Data = self.TF_Regex_Pos(Hasil_Data, politik, ekonomi, teknologi)
                self.IDF_Regex_Pos(Hasil_Data, fitur_all_doc)
            Hasil_Proses_Berita.append(fitur_all_doc)
            return Hasil_Proses_Berita, fitur_all_doc
        else :
            Hasil_Proses_Berita = []
            for berita in data.Berita:
                if(IT == True):
                    Hasil_Data = self.Indonesia_Tagger(berita, ct)
                if(TFR == True):
                    Hasil_Data = self.TF_Regex_Pos(Hasil_Data, politik, ekonomi, teknologi)
                Hasil_Proses_Berita.append(Hasil_Data)
            return Hasil_Proses_Berita, 0

    def Outputan_Data(self):
        for i in range(0, len(self.data_hasil['berita'])):
            print(str(i+1))
            print(self.data_hasil['berita'][i])     
            print('================================================')
        return None

    def __init__(self, data, IT, TFR, IDFR): #iniisialisasi data tagger
        ct = CRFTagger()
        ct.set_model_file('./library/tagger/indonesian_tagger')
        politik = ['politik', 'capres', 'cawapres', 'kebijakan', 'presiden', 'wakil presiden', 'pemerintah', 'kementrian',
               'politikus', 'menteri', 'partai', 'pemilihan umum', 'pemilu', 'dpr', 'pemkot', 'pemprov', 'kampanye']
        ekonomi = ['ekonomi', 'saham', 'pasar saham', 'pendapatan', 'investasi', 'suku bunga', 'bunga', 'tarif pajak', 
                   'keuangan', 'anggaran', 'pajak', 'miskin', 'biaya']
        teknologi = ['game', 'platform', 'versi', 'teknologi', 'programer', 'designer', 'jaringan', 'performa', 'prangkat lunak', 
                     'perangkat keras', 'fitur', 'smartphone', 'rilis', 'devloper', 'upgrade', 'fasilitas']
        
        #Proses tagger
        self.hasil_tagger, self.fitur_all_doc = self.Proses_Regex(data, IT, ct, TFR, politik, ekonomi, teknologi, IDFR)
        
        self.data_hasil = {}
        self.data_hasil['berita'] = self.hasil_tagger
        
    def Get_Hasil(self):
        return self.hasil_tagger
    
    def Get_Fitur_All_Doc(self):
        return self.fitur_all_doc
```

# Metode TF - IDF 

```python
class Tf_Idf():
    
    # ambil unique words
    def Bag_of_Word(self, data):
        bow = []
        for fitur in data:
            bow.append(fitur)
        return bow
    
    # bikin table term frequency
    def TF_Formula(self, bow, fitur_onedoc):
        tf_table = np.zeros((len(fitur_onedoc), len(bow)), dtype=int)
        for no_doc, doc in enumerate(fitur_onedoc):
            for no_fitur, fitur in enumerate(bow):
                # print(fitur)
                if(fitur in doc):
                    tf_table[no_doc,no_fitur] = doc[fitur]
        return tf_table
    
    # Calculate buat table idf
    def IDF_Formula(self, all_fitur, size_doc):
        for fitur in all_fitur:
            all_fitur[fitur] = math.log10(size_doc/all_fitur[fitur])
        return all_fitur
    
    def TF_IDF_Formula(self, tf, idf, bow):
        tfidf = np.zeros((len(tf), len(idf)), dtype=float)
        for i in range(len(tf)):
            for j, fitur in enumerate(bow):
                tfidf[i,j] = tf[i,j]*idf[fitur]
        return tfidf
    
    def Proses_Tf_Idf(self, fitur_all_doc, fitur_one_doc, BOW, TF, IDF, TFIDF): #proses tagger
        Hasil_Proses_Berita = []
        if(BOW == True):
            Hasil_Bow = self.Bag_of_Word(fitur_all_doc)
            Hasil_Data = Hasil_Bow 
        if(TF == True):
            Hasil_Tf = self.TF_Formula(Hasil_Data, fitur_one_doc)
            Hasil_Data = Hasil_Tf
        if(IDF == True):
            Hasil_Idf = self.IDF_Formula(fitur_all_doc, len(fitur_one_doc))
            Hasil_Data = Hasil_Idf
        if(TFIDF == True):
            Hasil_Data = self.TF_IDF_Formula(Hasil_Tf, Hasil_Idf, Hasil_Bow)
        if(TFIDF == True):
            return Hasil_Data
        else:
            Hasil_Proses_Berita.append(Hasil_Data)
            return Hasil_Proses_Berita
    
    def Save_Excel(self, berita, nfile):
        df = pd.DataFrame(berita)
        df.to_csv(nfile,  sep=',', encoding='utf-8', index=False)

    def Outputan_Data(self):
        for i in range(0, len(self.data_hasil['berita'])):
            print(self.data_hasil['berita'][i])     
            
        return None

    def __init__(self, fitur_all_doc, fitur_one_doc, BOW, TF, IDF, TFIDF): #iniisialisasi data tagger
        result_tfidf = './dummy/dummy_Hasil_Tf_Idf.csv'
        #Proses TF IDF
        hasil_tf_idf = self.Proses_Tf_Idf(fitur_all_doc, fitur_one_doc, BOW, TF, IDF, TFIDF)
        
        self.data_hasil = {}
        self.data_hasil['berita'] = hasil_tf_idf
        if(TFIDF == True):
            self.filepath = self.Save_Excel(hasil_tf_idf, result_tfidf)
```

# Cross Validation

```python
from sklearn.model_selection import KFold

result_tfidf    = './dummy/dummy_Hasil_Tf_Idf.csv'
data_label      = './dummy/dummy_label.csv'
path_hasil_tfidf = './dummy/Report_tfidf.txt'
metode_tfidf    = 'TF-IDF'
```

```python
df = pd.read_csv(result_tfidf)  # tabel hasil ekstraksi
label = pd.read_csv(data_label) # tabel label sesuai data ekstraksi
X = df.values                   # tabel ekstraksi dijadikan array biar mudah

Politik = label['Politik'].values           # array label anjuran
Ekonomi = label['Ekonomi'].values         # array label larangan
Teknologi = label['Teknologi'].values       # array label informasi
arr_label = [Politik, Ekonomi, Teknologi]  # gabungan dari 3 array label

label_pred = [] # inisialisasi array untuk prediksi label
label_real = [] # inisialisasi array untuk label yang sesungguhnya
indeks_tes = []
kf = KFold(n_splits=10, shuffle=False) # data di bagi 5 fold dan 1x perulangan
jeniskelas = ['POLITIK','EKONOMI','TEKNOLOGI']
```

# Klasifikasi Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

def klasifikasi_random(X_train, y_train, X_test):
    classifier = RandomForestClassifier(n_estimators = 200, criterion = 'entropy', random_state = 0)
    classifier.fit(X_train,y_train)
    classifier_predictions = classifier.predict(X_test)
    
    return classifier_predictions
```

```python
for c, y in enumerate(arr_label): # untuk setiap label (ada length = 3)
    single_label_pred = [] # inisialisasi array lebel_pred perlabel
    single_label_real = [] # inisialisasi array lebel_real perlabel
    indeks_tesperfold = []
    count = 0
    for train_index, test_index in kf.split(X): # looping untuk setiap fold yang telah dibagi menjadi 5
        count += 1
        print('FOLD KE-', count , ' | KELAS-', jeniskelas[c], ' | METODE-', metode_tfidf)
        print("TEST:",test_index)
#         print("---------------------------------------------------")
        X_train, X_test = X[train_index], X[test_index] 
        y_train, y_test = y[train_index], y[test_index]

        # PROSES KLASIFIKASINYA DISINI
        predictions = klasifikasi_random(X_train, y_train, X_test)
        # ============================
#         print("X_test = ", X_test)
#         print("y_test = ", y_test)
        print (predictions, ' --- ', y_test)
#         print("---------------------------------------------------")
#         print("---------------------------------------------------")
        single_label_pred.append(predictions)
        single_label_real.append(y_test)
        indeks_tesperfold.append(test_index)
    label_pred.append(single_label_pred)
    label_real.append(single_label_real)
    indeks_tes.append(indeks_tesperfold)
```

# Harming Loss

```python
def hamming_loss(label_pred, label_real, len_data, jum_label):
    count_salah = 0
    for i, pred in enumerate(label_pred):
        for j, angka in enumerate(pred):
            if label_pred[i][j] == label_real[i][j]:
                continue
            else:
                count_salah += 1
    hamming_loss = (1/len_data)*(1/jum_label)*count_salah
    return hamming_loss

def save_info_kfold(label_pred, label_real, indeks_tes, hasil_hammingloss, output_path):
    out = open(output_path, 'w')
    jeniskelas = ['POLITIK','EKONOMI','TEKNOLOGI']
    for n, elemen in enumerate(indeks_tes):
        out.write('# KELAS ' + jeniskelas[n]  + ' # \n')
        for k, i in enumerate(elemen):
            out.write('- Fold Ke ' + str(k+1) + ' - \n')
            out.write('Indeks Data test : ')
            for j in i:
                out.write('{} '.format(j))
            out.write('\n')
            out.write('Label True       : ')
            for l in label_real[n][k]:
                out.write('{} '.format(l))
            out.write('\n')
            out.write('Label Prediksi   : ')
            for m in label_pred[n][k]:
                out.write('{} '.format(m))
            out.write('\n')
            out.write('Hamming Loss ke-'+ str(k+1) + ': ' +  str(hasil_hammingloss[k]) + '\n')
        out.write('\n')
        n_HLoss = np.mean(hasil_hammingloss)
    out.write('====================================== \n ')
    out.write('HAMMING LOSS     : ' + '{} '.format(n_HLoss) + '\n')
    out.write('====================================== ')
    out.close()
```

```python
hasil_hammingloss = []
for i in range(kf.get_n_splits()):
    label_prediksi = []
    label_benar    = []
    for j in range(len(label_pred)):
        label_prediksi.append(label_pred[j][i])
        label_benar.append(label_real[j][i])
    # INI PERHITUNGAN HAMMING LOSSNYA
    nilai_hl = hamming_loss(label_prediksi, label_benar, len(label_real[0][0]), len(arr_label))
    # ========================
    hasil_hammingloss.append(nilai_hl)
#        print('Hamming Loss K ke', i+1 ,': ' , nilai_hl)
n_rata_hl = np.mean(hasil_hammingloss)
#    print('Hamming Loss Keseluruhan : ' , n_rata_hl)

save_info_kfold(label_pred, label_real, indeks_tes, hasil_hammingloss, path_hasil_tfidf)

print(n_rata_hl)
```
